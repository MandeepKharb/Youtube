```python
#importing dependencies
import pandas as pd
import numpy as np
import requests
import json
from bs4 import BeautifulSoup
import csv
import os
import shutil
```

```python
# Retrieving Food List
food = pd.read_csv(os.path.join("..","static","food_list", "food_list.csv"))
food

```

```python
#Deleting and creating Directiories
shutil.rmtree(os.path.join("images","valid"))
shutil.rmtree(os.path.join("images","test"))
shutil.rmtree(os.path.join("images","train"))
os.mkdir(os.path.join("images","valid"))
os.mkdir(os.path.join("images","test"))
os.mkdir(os.path.join("images","train"))
```

```python
# creating folder structure
for i in food:
    os.mkdir(os.path.join("images","train",i))
    os.mkdir(os.path.join("images","valid",i))
    os.mkdir(os.path.join("images","test",i))
```

```python
#Downloading Train Data Sets
with open("C:\\Users\\mandeep.singh\\Machine-Learning-project-master\\static\\food_list\\food_list.json", "r", encoding="utf8") as f:
    food_translate = json.load(f)

# download images to train dir
NUM_IMAGES = 5
images_per_group = NUM_IMAGES

for food in food_translate.keys():
    # check number of files in directory
    for r, d, files in os.walk(os.path.join("images", "train", food)):
        files_in_directory = len(files)

    if files_in_directory < images_per_group:
        print("Collecting image urls for {}.".format(food))

        food_name = food_translate[food]
        page = 1
        image_urls = []
        while len(set(image_urls)) <= images_per_group:
            # download images from Yahoo             
            #yahoo_url = "https://in.images.search.yahoo.com/search/images?p="+food_name+"&ei=UTF-8&b="+str(page)
            yahoo_url = "https://www.google.com/search?safe=off&site=&tbm=isch&source=hp&q={q}&oq={q}&gs_l=img"
            yahoo_url.format(q=food_name)
            print(yahoo_url.format(q=food_name))
            response = requests.get(yahoo_url,verify=False)
            print(response.status_code)
            soup = BeautifulSoup(response.text, "lxml")
           # print(soup)
            image_urls.extend([i["src"] for i in soup.findAll("img")[:-1]])
           # print(image_urls)
            page += 20

        print("Downloading images for {}.".format(food))
        for index, img_url in enumerate(set(image_urls)):
            img_data = requests.get(img_url,verify=False).content
            with open(os.path.join("images", "train", food, food+"_"+str(index)+".jpg"), "wb") as handler:
                handler.write(img_data)
    else:
        print("Images for {} is already downloaded.".format(food))
```

```python
# !find . -name '.DS_Store' -type f -delete
#Splitting Random images to Test & Valid dir
train = os.path.join("images", "train")
valid = os.path.join("images", "valid")
test = os.path.join("images", "test")
    
# transfer from train to test 80-20
train_dir = os.listdir(train)

#for i in train_dir:
#    os.mkdir(os.path.join("images","valid",i))
#    os.mkdir(os.path.join("images","test",i))
    
for idx in train_dir:
    num_transfers = np.ceil(len(os.listdir(os.path.join(train, idx)))*0.2).astype("int")
    print(num_transfers)
    files_to_transfer = np.random.choice(os.listdir(os.path.join(train, idx)), num_transfers, replace=False)
    for file in files_to_transfer:
            shutil.move(os.path.join(train, idx, file), os.path.join(test, idx))

# transfer from train to valid 80-20
train_dir = os.listdir(train)
for idx in train_dir:
    num_transfers = np.ceil(len(os.listdir(os.path.join(train, idx)))*0.2).astype("int")
    files_to_transfer = np.random.choice(os.listdir(os.path.join(train, idx)), num_transfers, replace=False)
    for file in files_to_transfer:
            shutil.move(os.path.join(train, idx, file), os.path.join(valid, idx))
```

```python
# revert folder image structure
# change source and dest(ination)
revert = os.listdir(valid)
for idx in revert:
     files = os.listdir(os.path.join(valid, idx))
     for file in files:
         shutil.move(os.path.join(valid, idx, file), os.path.join(train, idx, file))
```

```python
revert = os.listdir(test)
for idx in revert:
     files = os.listdir(os.path.join(test, idx))
     for file in files:
         shutil.move(os.path.join(test, idx, file), os.path.join(train, idx, file))
```
