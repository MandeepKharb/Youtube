```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

import os
from os import listdir
from os.path import isfile, join
from keras.applications.inception_v3 import preprocess_input
from keras.models import load_model
```

```python
root_dir = 'C:\\Users\\mandeep.singh\\Machine-Learning-project-master\\Classification_Models\\images\\train\\'
sorted_food_dirs = sorted(os.listdir(root_dir))
sorted_food_dirs=sorted_food_dirs[1:]
sorted_food_dirs
from keras.preprocessing import image
name=[]
picture=[]
for i in sorted_food_dirs:
    food_dir = i
    all_files = os.listdir(os.path.join(root_dir, food_dir))
    #print(all_files)
    j=0
    while j<10:
        rand_img = np.random.choice(all_files)
        path=os.path.join(root_dir, food_dir, rand_img)
        print(path)
        img = image.load_img(path,color_mode = "grayscale")
        x=image.img_to_array(img)
        picture.append(x) 
        name.append(food_dir)
        j=j+1
#         if x.shape[0]==512:
#             if x.shape[1]==512:
                
len(picture)
```

```python
import pandas as pd
Food_matrix=pd.DataFrame({"name":name,"picture":picture})
```

```python
X=Food_matrix.picture
y=Food_matrix.name
X[1].shape[1]*X[1].shape[2]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
label_encoder.fit(y)
encoded_y=label_encoder.transform(y)
```

```python
type(X)
```

```python
for i in range(0,len(X)):
    ndims=X[i].shape[1]*X[i].shape[2]
    X[i]=X[i].reshape(X[i].shape[0],ndims).reshape(-1,1)   
```

```python
for i in range(0,len(X)):
    ndims=X[i].shape[0]*X[i].shape[1]
    X[i]=X[i].reshape(ndims)
```

```python
X.shape
```

```python
X=preprocess_input(X)
```

```python
X_list=[]
for i in range(0,len(X)):
    X_list.append(X[i].tolist())    
len(X_list)
```

```python
scores=[]
from sklearn.neighbors import KNeighborsClassifier
for k in range(3,10,2):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_list,encoded_y)
    score=knn.score(X_list,encoded_y)
    scores.append(score)
    print(f"k:{k}, score:{score}")
```

```python
K=list(range(3,10,2))
plt.plot(K,scores,marker="*")
plt.xlabel("K neighbors")
plt.ylabel("Accuracy score")
plt.show()
```

```python
#The best result is for k=3 score=0.4
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_list,encoded_y)
```

```python
import joblib
filename = 'KNN_best_model.sav'
joblib.dump(knn, filename)
```
