# Gradient Descent Algorithm and How it works 

```python
### idea behind Gradient Descent Algo is to find the best value of m and b in if the linear eqn is 
### is of this form -> y = mx + b where m is slope and b is intercept 
### and we will do it iterations , while picking some starting value of m and b then keep doing iteration 
### and at each iteration we will find value of cost function, when the cost function will be minimum 
### then we will pick the value of m and b, which will generate the best fit line

### cost function we will use here is is MSE (mean squared error)


```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
```

```python
y = np.array([5,7,9,11,13])
```

```python
# below function is for gradient_descent
def gradient_descent(x,y):
    m_curr = b_curr = 0;  ## intial values of m and b (slope and intercept)
    iterations = 1000;  ## number of iteration we want to keep checcking cost function
    n = len(x) # number of data points
    learning_rate = 0.001
    
    for i in range(iterations):
        y_pred = m_curr*x + b_curr
        cost = (1/n)*sum(val**2 for val in (y-y_pred)) ## cost function(mean sqaured error), its formula can be seen over internet 
        m_derivative = -(2/n)*sum(x*(y-y_pred)) ## partial derivative of cost function wrt m (slope), which will give us direction
        b_derivative = -(2/n)*sum(y-y_pred)  ## partial derivative of cost function wrt b
        m_curr = m_curr - learning_rate * m_derivative
        b_curr = b_curr - learning_rate * b_derivative 
        print("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost,i))
```

```python
gradient_descent(x,y) ## gradient_descent function is called and see cost function is decreasing at each iteration
## we can play around with the values of iterations and learning rate , ideally learning rate should be low and 
## iteration should be high, so that model could learn and find the best fit line by minimizing the cost function
```
