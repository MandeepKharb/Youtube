```python
import pandas as pd
from patsy import dmatrices
from matplotlib import pyplot as plt
import numpy as np
 
```

```python
df = pd.read_csv('PowerPlant.csv')
df.head()
```

```python
df.plot.scatter(x='Ambient Temperature', y='Power Output')
plt.xlabel('Ambient Temperature', fontsize=18)
plt.ylabel('Power Output', fontsize=18)
plt.show()
```

```python
df.plot.scatter(x='Volume', y='Power Output')
plt.xlabel('Volume', fontsize=18)
plt.ylabel('Power Output', fontsize=18)
plt.show()
```

```python
df.plot.scatter(x='Ambient Pressure', y='Power Output')
plt.xlabel('Ambient Pressure', fontsize=18)
plt.ylabel('Power Output', fontsize=18)
plt.show()
```

```python
df.plot.scatter(x='Relative Humidity', y='Power Output')
plt.xlabel('Relative Humidity', fontsize=18)
plt.ylabel('Power Output', fontsize=18)
plt.show()
```

```python
df.corr()['Power Output']
```

# Ambient Temperature has LINEAR RELATIONSHIP with Power Output
